Para cada elemento do indice da lista de carteiras é calculada a distância do elemento de entrada que deve ser classificado. Se as listas de distância ou de resposta (dist_k e resp_k) tiver menos que K elementos, serão apenas adicionados. Se tiverem K elementos, será verificado o maior valor da lista de distancia (dist_k) e o indice associado a ele. Para verificação da moda, é criado um dicionário onde as chaves são as definições de classificação e os valores são a contagem de cada classificação nas listas de dist_k e resp_k. Caso só exista UM valor máximo, será retornada a classificação que mais se repete.<br>
<br>
- E se a quantidade de valores de classificação resultar em empate? Ex: 2x "Agressivo" e 2x "Moderado"? ou 2x "Agressivo", 2x "Moderado" e 2x "Conservador"<br>
Para esse caso o dicionário de decisão passa a ser somente o relacionado aos valores que mais se repetem e então é criado um segundo dicionário onde as chaves são novamente as definições de classificação, mas os valores são a MENOR distância euclidiana de cada uma das classificações. Nesse caso, é retornada a classificação de MENOR distância.<br>
<br>
- E se as distâncias euclidianas de duas ou mais classificações forem as mesmas?<br>
Nesse caso o ideal é verificar com a equipe de negócio qual o melhor plano e estratégia. Para esse caso em específico, Assumi que o melhor resultado seria o que trouxesse o menor risco para o cliente, considerando que devem ser ofertados produtos e ativos de menor risco. Caso o perfil do cliente seja de maior risco, é mais provável que uma segunda verificação ocorra conforme a carteira de ativos financeiros vá sendo atualizada e incrementada.

```python
def separar_dados(dados):
    carteiras = []
    classificacao = []

    # De acordo com o indice de cada elemento do argumento informado
    for i in range(len(dados)):
        
        # Adiciona o valor de classificação
        classificacao.append(dados[i][1])
        
        # Adiciona a coleção de carteira
        carteiras.append(dados[i][2])

    # Retorna as listas
    return classificacao, carteiras
```

```python
def dist_euclidiana(valor_base, valor_input):

    euclid = 0

    # Caso seja passado uma coleção (Tupla ou Lista)
    if isinstance(valor_base, (tuple)) and isinstance(valor_input, (tuple)) or isinstance(valor_base, (list)) and isinstance(valor_input, (list)):

        # Verificação se as coleções possuem o mesmo tamanho
        if len(valor_base) == len(valor_input):

            # Cálculo do valor quadrático da distância
            for valor in zip(valor_base, valor_input):
                euclid += (valor[0] - valor[1])**2
        
        # Se não possuem o mesmo tamanho, retorna erro
        else:
            return TypeError(f"As coleções devem possuir o mesmo tamanho. As informadas possuem tamanhos {len(valor_base)} e {len(valor_input)}")
    
    # Caso seja passado um número (float ou int)
    elif isinstance(valor_base, (int, float)) and isinstance(valor_input, (int, float)):

        # Cálculo do valor quadrático da distância
        euclid += (valor_base - valor_input)**2

    # Se não possuem os tipos permitidos, retorna erro
    else:
        return TypeError(f"Deve ser informado o mesmo tipo de coleção ou dados numéricos. Atualmente se tem {type(valor_base)} e {type(valor_input)}")
    
    # Retorna o valor da distância
    return euclid ** (1/2)
```

```python
def decisao_knn(lista_distancia, lista_resposta):
    if len(lista_resposta) == 0:
        print(f"Situações com erro não foram resolvidas.")

    else:
        # Se o valor de resposta for int ou float, retorna a média
        if isinstance(lista_resposta[0], (int, float)):

            regressao = sum(lista_resposta) / len(lista_resposta)

            return regressao

        # Se o valor de resposta for uma string, retorna moda
        elif isinstance(lista_resposta[0], str):
            # Dicionário com as chaves sendo a classificação
            dicionario_count = {}
            dicionario_count = dicionario_count.fromkeys(lista_resposta)
            
            # Valores das chaves é a contagem das vezes que aparecem na lista de resposta
            for chave in dicionario_count:
                dicionario_count[chave] = lista_resposta.count(chave)
            
            # Se o maior valor aparecer apenas uma vez nos valores do dicionário, retorna a chave como resposta
            if list(dicionario_count.values()).count(max(dicionario_count.values())) == 1:
                classificacao = list(dicionario_count.keys())[list(dicionario_count.values()).index(max(list(dicionario_count.values())))]
            
            # Se o maior valor aparecer mais de uma vez, alterar o dicionário para que tenha apenas as chaves com maior valor
            else:
                dicionario_count = {chave_count:dicionario_count[chave_count] for chave_count in dicionario_count if dicionario_count[chave_count] == max(dicionario_count.values())}
                dicionario_dist = {}

                # Criar um dicionário com as chaves que mais se repetem e os valores são a MENOR distância euclidiana relativa a aquele valor nas listas de resposta
                dicionario_dist = dicionario_dist.fromkeys(dicionario_count.keys(), max(lista_distancia))

                for i in range(0, len(lista_distancia)):
                    if lista_resposta[i] in dicionario_dist.keys():
                        if dicionario_dist[lista_resposta[i]] > lista_distancia[i]:
                            dicionario_dist[lista_resposta[i]] = lista_distancia[i]

                # Se existir apenas UM valor mínimo no dicionário de distância, retorna a chave como resposta
                if list(dicionario_dist.values()).count(min(dicionario_dist.values())) == 1:
                    classificacao = list(dicionario_dist.keys())[list(dicionario_dist.values()).index(min(list(dicionario_dist.values())))]

                # Se existirem valores repetidos de distância euclidiana mínima, retorna o perfil com menos investimentos de risco
                else:
                    if "Conservador" in dicionario_count.keys():
                        classificacao = "Conservador"
                    
                    else:
                        classificacao = "Moderado"

            return classificacao
```

```python
def projeto_knn(lista_resposta, lista_valores, valor_input, k=5):

    # Loop e Armazenamento de k Itens mais próximos
    dist_k = []
    resp_k = []

    for indice in range(len(lista_valores)):

        # Distância Euclidiana e manutenção de listas com k itens
        distancia = dist_euclidiana(lista_valores[indice], valor_input)
        
        # Se o retorno da distância não for um erro
        if type(distancia) != TypeError:

            if len(dist_k) < k:
                dist_k.append(distancia)
                resp_k.append(lista_resposta[indice])
            
            else:
                if max(dist_k) > distancia:
                    remover = dist_k.index(max(dist_k))
                    dist_k.pop(remover)
                    resp_k.pop(remover)
                    
                    dist_k.append(distancia)
                    resp_k.append(lista_resposta[indice])
        
        # Se o retorno for um erro, interromper o processo
        else:
            print(f"{distancia}")
            break
    
    # Decisão
    return decisao_knn(dist_k, resp_k)
```

```python
lista_prevista = []
lista_real = []

# Definição do valor de k
valor_k = 8

# Alimentação das listas de valores reais e previstos
for i in range(len(data)):
    lista_prevista.append(projeto_knn(resposta_base, valores_base, data[i][2], k=valor_k))
    lista_real.append(data[i][1])

acerto = 0
erro = 0
indices = []

# Comparação e métricas de avaliação
for i in range(len(lista_prevista)):
    if lista_prevista[i] == lista_real[i]:
        acerto += 1
    else:
        erro += 1
        indices.append(i)

acuracia = acerto / len(lista_prevista)

print(f"Tamanho da lista: {len(lista_prevista)}")
print(f"Previsões corretas: {acerto}")
print(f"Previsões erradas: {erro}")
print(f"Indices das previsões erradas: {indices}")
print(f"Acurácia do modelo: {acuracia}")
    
```
